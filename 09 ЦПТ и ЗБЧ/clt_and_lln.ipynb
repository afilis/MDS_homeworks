***Центральная предельная теорема***  
В этом задании нужно убедиться, что ЦПТ действительно работает.  
Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

a = 1
b = 4
rv = sts.uniform(a, b - a)
sample = rv.rvs(size=1000)
x = np.linspace(0, 5, 1000)
pdf = rv.pdf(x)
plt.title("Равномерное распределение на отрезке")
plt.hist(sample, density=True, bins=30, label="Гистограмма выборки")
plt.plot(x, pdf, label="Теоретическая плотность распределения", alpha=0.5)
plt.legend()
plt.ylabel("Доля в выборке, f(x)")
plt.xlabel("x")
plt.show()
```

```python
def score_hist_data(rv, samples_count, n):
    a = np.array([])
    for number in range(samples_count):
        sample = rv.rvs(size=n)
        a = np.append(a, sample.mean())
    return a

def plot_hist(plt, rv, samples_count, n, bins=30):
    a = score_hist_data(rv, samples_count, n)
    plt.hist(a, density=True, bins=bins, label=f"Средние выборок размера n = {n}")

samples_count = 1000
for n in [2, 5, 10, 30, 50]:
    plot_hist(plt, rv, samples_count, n)

plt.title("Гистограммы средних выборок в зависимости от размера выборок")
plt.legend()
plt.ylabel("Средние выборок")
plt.xlabel("x")
plt.show()
print("""По гистограмме видно, что чем больше объем выборки n, тем ближе распределение к нормальному.
Это отражает справедивость центральной предельной теоремы.""")
```
