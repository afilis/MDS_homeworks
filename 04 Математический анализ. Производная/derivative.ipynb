**Уровень 1**  
Постройте график первой и второй производной функции: $f(x) = (sin{2x+1})^5$ на отрезке [-5; 5].

```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt

def f(x):
    return (np.sin(2 * x + 1)) ** 5

x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]

plt.plot(x, fx)
plt.plot(x, f1x)
plt.plot(x, f2x)
plt.grid(b=True, which='major')
plt.title("График функции f(x) = (sin2x+1)^5 и ее 1-ой и 2-ой производных")
plt.legend(["f (x)", "f '(x)", "f ''(x)"])
plt.show()
```

**Уровень 0**  
Посчитайте значение 1-ой и 2-ой производной фунции $f(x) = x^5 + 4sin(2x) + cos(3x+3)$ в точке x = 1.

```python
def f(x):
    return x ** 5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)

x = 1
fx = f(x)
f1x = derivative(f, x, dx=1e-6, n=1)
f2x = derivative(f, x, dx=1e-6, n=2)

print(f"Значение функции в точке x = {x} равно {fx}.")
print(f"Значение 1-ой производной в точке x = {x} равно {f1x}.")
print(f"Значение 2-ой производной в точке x = {x} равно {f2x}.")
```
